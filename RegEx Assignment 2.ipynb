```python
# question 3

the_string = 'this will find all words of atleast four characters in the string'

string_pattern = r"\w{4}"

re_pattern = re.compile(string_pattern)

find_4words = re_pattern.findall(the_string) # this will find all words of atleast four characters in the string
find_4words
```
